```python
# Source: 
# https://github.com/chasinginfinity/ml-from-scratch/tree/master/03%20Linear%20Regression%20in%202%20minutes
from sklearn.linear_model import LinearRegression

data = pd.read_csv('data.csv')  # load data set
X = data.iloc[:, 0].values.reshape(-1, 1)  # values converts it into a numpy array
Y = data.iloc[:, 1].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
```

```python
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(X, Y)  # perform linear regression 


# linear_regressor.get_params()
linear_regressor.coef_
Y_pred = linear_regressor.predict(X)  # make predictions


plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
print(linear_regressor.coef_)
plt.show()
```

```python
# Ridge regresija
from sklearn.linear_model import Ridge
linear_regressor = Ridge()

linear_regressor.fit(X, Y)  # perform linear regression 


# linear_regressor.get_params()
linear_regressor.coef_
Y_pred = linear_regressor.predict(X)  # make predictions


plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
print(linear_regressor.coef_)
plt.show()
```

```python
# Laso regresija, var paspēlēties ar 'aplha'  vērtību no 0.1 līdz 5
from sklearn.linear_model import Lasso
linear_regressor = Lasso()

linear_regressor.fit(X, Y)  # perform linear regression 


# linear_regressor.get_params()
linear_regressor.coef_
Y_pred = linear_regressor.predict(X)  # make predictions


plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
print(linear_regressor.coef_)
plt.show()
```
